```python
A = int(input("enter the number "))
#Independent Variable
X = [8, 12, 10]
#Dependent Variable
Y = [10, 16, 13]

# Calculating mean
def mean(values):
    return sum(values) / len(values)
# Calcualting Deviation
def deviation():
    mean_X = mean(X)
    mean_Y = mean(Y)

    list_of_dev_X = [x - mean_X for x in X]
    list_of_dev_Y = [y - mean_Y for y in Y]

    product_of_dev = [list_of_dev_X[i] * list_of_dev_Y[i] for i in range(len(X))]
    squared_dev_X = [dev ** 2 for dev in list_of_dev_X]

    return sum(product_of_dev), sum(squared_dev_X)

#Calculating Slope[How much Y changes upon a unit change in X]
def slope():
    numerator, denominator = deviation()
    return numerator / denominator

#Calculating Intercept[Value of Y when X is 0]
def intercept():
    mean_X = mean(X)
    mean_Y = mean(Y)
    m = slope()
    b = mean_Y - m * mean_X
    return b, m

# Prediction for new value
def predict(A):
    b, m = intercept()
    return m * A + b

# Outputs
print("Slope (m):", slope())
b, m = intercept()
print("Intercept (b):", b)
print(predict(A))
```
